```python
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])  # Sequence of features
        y.append(data[i+sequence_length, 0])  # Select only one target (first column)
    return np.array(X), np.array(y)
```

```python
from sklearn.preprocessing import StandardScaler
import joblib
import numpy as np

sequence_length = 3
z_threshold = 4 # Define the Z-score threshold for outlier removal

# Define numerical & categorical features
numerical_features = ['discounted_price', 'listPrice']
categorical_features = ['website', 'month']

# Dictionary to store scalers for each category
scalers = {}

# Placeholder for processed data
all_data = []

# Grouping by category_id
grouped_by_category = df.groupby('category_id')

for category_id, group in grouped_by_category:
    # Extract numerical and categorical data separately
    numerical_data = group[numerical_features].values
    categorical_data = group[categorical_features].values  # Keep as is

    # 1 Remove outliers using Z-score
    mean = np.mean(numerical_data, axis=0)
    std = np.std(numerical_data, axis=0)
    z_scores = np.abs((numerical_data - mean) / std)

    # Mask to filter out outliers
    mask = (z_scores < z_threshold).all(axis=1)

    outliers_removed = group[~mask]

    # Print removed outliers (optional: save to CSV for review)
    if not outliers_removed.empty:
        print(f" Outliers removed from category {category_id}:")
        print(outliers_removed)
        # outliers_removed.to_csv(f"outliers_category_{category_id}.csv", index=False)

    numerical_data_filtered = numerical_data[mask]
    categorical_data_filtered = categorical_data[mask]  # Keep matching categorical data

    # 2 Standard Scale numerical features only
    scaler = StandardScaler()
    numerical_scaled = scaler.fit_transform(numerical_data_filtered)

    # Save the scaler for this category
    scalers[category_id] = scaler
    joblib.dump(scaler, f"{category_id}_scaler.pkl")

    # 3 Concatenate scaled numerical data with unscaled categorical features
    processed_data = np.hstack((numerical_scaled, categorical_data_filtered))

    # Add category_id column as is (no scaling)
    category_id_column = np.full((processed_data.shape[0], 1), category_id)
    processed_data_with_category = np.hstack((processed_data, category_id_column))

    # 4 Create sequences
    X, y = create_sequences(processed_data_with_category, sequence_length)

    # Store processed data
    all_data.append((X, y))
```

```python
from sklearn.model_selection import train_test_split

# Flatten all the sequences and targets from all_data
X_all = np.concatenate([x for x, y in all_data], axis=0)
y_all = np.concatenate([y for x, y in all_data], axis=0)

# Split data into training and validation sets
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, BatchNormalization
from tensorflow.keras.optimizers import Adam

# Build the complex model
model = Sequential()

# Add a Bidirectional LSTM layer
model.add(Bidirectional(LSTM(units=128, return_sequences=True), input_shape=(X_train.shape[1], X_train.shape[2])))
# model.add(LSTM(units=128, return_sequences=True), input_shape=(X_train.shape[1], X_train.shape[2]))
model.add(Dropout(0.2))  # Increased dropout rate for regularization

# Add another Bidirectional LSTM layer
model.add(Bidirectional(LSTM(units=64, return_sequences=True)))
# model.add(LSTM(units=64, return_sequences=True))
model.add(Dropout(0.2))

# Add a third LSTM layer with return_sequences=False
model.add(LSTM(units=64, return_sequences=False))
model.add(Dropout(0.2))

# Add Batch Normalization for improved training stability
model.add(BatchNormalization())

# Add more Dense layers
model.add(Dense(128, activation='relu'))  # Increased units in Dense layer
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))  # Added more layers with different units
model.add(Dropout(0.2))

# Add output layer
model.add(Dense(1))  # Regression output (no activation)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

model_checkpoint = ModelCheckpoint(
    filepath='best_model.keras',  # Save the best model to this file
    monitor='val_loss',  # Track validation loss
    save_best_only=True,  # Save only the best model (lowest val_loss)
    mode='min',  # Lower validation loss is better
    verbose=1  # Print messages when a better model is saved
)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mae'])

# Summary of the model
model.summary()
```

```python
# Train the model with EarlyStopping
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_val, y_val), callbacks=[early_stopping, model_checkpoint])

# history = model.fit(X_train, y_train, epochs=50, batch_size=32, callbacks=[early_stopping])
```

```python
from datetime import datetime
import numpy as np
import joblib
import pandas as pd
import os

# Website mapping
website_map = {"Amazon": 0, "Flipkart": 1, "Snapdeal": 2}

# Define feature lists
numerical_features = ['discounted_price', 'listPrice']
sequence_length = 3

def load_scaler(category_id):
    """ Load the scaler for the given category_id. """
    scaler_path = f"{category_id}_scaler.pkl"
    if os.path.exists(scaler_path):
        return joblib.load(scaler_path)
    return None

def preprocess_input(data, month):
    """ Preprocess input data for model prediction. """
    df = pd.DataFrame([data])

    # Remove last 2 characters from ASIN
    df["asin"] = df["asin"].str[:-2]

    # Map website to numerical values
    df["website"] = df["website"].map(website_map).fillna(-1).astype(int)

    df.at[0, "month"] = month

    # Keep only relevant columns
    df = df[['listPrice', 'website', 'month', 'category_id', 'price_history']]

    # Convert price history into a DataFrame
    price_df = pd.DataFrame(df["price_history"].iloc[0])
    price_df["date"] = pd.to_datetime(price_df["date"], errors="coerce")
    price_df["month"] = price_df["date"].dt.month

    # Map discounted prices for the selected month
    df["discounted_price"] = df["month"].map(price_df.set_index("month")["discounted_price"])

    df = df[['discounted_price', 'listPrice', 'website', 'month', 'category_id', 'price_history']]

    # Scale numerical features
    category_id = df["category_id"].values[0]
    scaler = load_scaler(category_id)
    if scaler:
        df[numerical_features] = scaler.transform(df[numerical_features])
    else:
        return None, "Unknown category_id"

    # Get last 3 months (handling wrap-around for January)
    last_3_months = [(month - i) if (month - i) > 0 else (month - i + 12) for i in range(2, -1, -1)]

    # Extract discounted prices for the last 3 months
    selected_prices = []
    for m in last_3_months:
        row = price_df[price_df["month"] == m]
        if not row.empty:
            selected_prices.append(row["discounted_price"].values[0])
        else:
            selected_prices.append(0)  # If no data, fill with 0

    selected_prices_scaled = []

    if scaler:
        dummy_array = np.zeros((len(selected_prices), 2))
        dummy_array[:, 0] = selected_prices
        # Apply the scaler to this dummy array
        selected_prices_scaled = scaler.transform(dummy_array)[:, 0]
    else:
        selected_prices_scaled = selected_prices

    # Create sequence data
    sequences = []
    for i in range(sequence_length):
        row = [
            selected_prices_scaled[i],  # discounted_price for that month
            df["listPrice"].values[0],
            df["website"].values[0],
            last_3_months[i],  # month
            df["category_id"].values[0],
        ]
        sequences.append(row)

    sequence_data = np.array(sequences)

    return np.expand_dims(sequence_data, axis=0), None  # Add batch dimension
```

```python
import re
from datetime import datetime

def convert_row_to_format(row):
    try:
        # Get the 'price_history' string
        price_history_str = row['price_history']

        # Pattern to match datetime objects
        datetime_pattern = r"datetime\.datetime\((.*?)\)"

        # Replace the datetime objects with ISO formatted strings
        def datetime_replacer(match):
            # Extract the datetime string inside the match
            datetime_str = match.group(1)
            # Convert it into datetime object
            dt_obj = datetime.strptime(datetime_str, "%Y, %m, %d, %H, %M")
            # Return the ISO format of the datetime
            return f'"{dt_obj.isoformat()}"'

        # Apply the regex to replace all datetime objects
        price_history_str = re.sub(datetime_pattern, datetime_replacer, price_history_str)

        # Now try to convert the modified string to a list of dictionaries
        price_history = eval(price_history_str)

    except Exception as e:
        print(f"Error converting 'price_history' for ASIN {row['asin']}: {e}")
        price_history = []

    # Construct the formatted row
    formatted_row = {
        "asin": row["asin"],
        "title": row["title"],
        "stars": row["stars"],
        "reviews": row["reviews"],
        "price": row["price"],
        "listPrice": row["listPrice"],
        "category_id": row["category_id"],
        "isBestSeller": row["isBestSeller"],
        "boughtInLastMonth": row["boughtInLastMonth"],
        "website": row["website"],
        "price_history": price_history  # Now correctly formatted
    }

    return formatted_row
```
